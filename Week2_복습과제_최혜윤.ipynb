```python
import torch
import torchvision.transforms as transforms
from torchvision.datasets import MNIST
from torch.utils.data import DataLoader, Subset
import random

# Tensor Î≥ÄÌôòÏùÑ ÏúÑÌï®
transform = transforms.Compose([
    transforms.ToTensor()
])

# MNIST Îç∞Ïù¥ÌÑ∞ÏÖã Î∂àÎü¨Ïò§Í∏∞
train_dataset = MNIST(root='./data', train=True, transform=transform, download=True)
test_dataset = MNIST(root='./data', train=False, transform=transform, download=True)

# Î¨¥ÏûëÏúÑÎ°ú ÏÉòÌîå Ïù∏Îç±Ïä§ ÏÑ†ÌÉù
train_indices = [i for i, label in enumerate(train_dataset.targets) if label in [0, 1, 2, 3, 4]]
test_indices = [i for i, label in enumerate(test_dataset.targets) if label in [0, 1, 2, 3, 4]]

# SubsetÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏÉòÌîåÎßÅÎêú Îç∞Ïù¥ÌÑ∞ÏÖã ÏÉùÏÑ±
train_subset = Subset(train_dataset, train_indices)
test_subset = Subset(test_dataset, test_indices)


# DataLoader ÏÉùÏÑ±
batch_size = 64  # Î∞∞Ïπò ÌÅ¨Í∏∞Îäî 64Î°ú ÏÑ§Ï†ï (batch_sizeÎ•º Îã§Î•¥Í≤å Ïã§ÌñâÌï¥Î≥¥ÏÖîÎèÑ Ï¢ãÏäµÎãàÎã§)
train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_subset, batch_size=batch_size, shuffle=False)
```

```python
# Îç∞Ïù¥ÌÑ∞ÏÖã ÌÅ¨Í∏∞ ÌôïÏù∏ÏùÑ ÏúÑÌï¥ Ìï¥Îãπ ÏÖÄÏùÑ Ïã§ÌñâÏãúÏºú Ï£ºÏÑ∏Ïöî
print(f"Train dataset size: {len(train_subset)}")
print(f"Test dataset size: {len(test_subset)}")
```

```python
# Îç∞Ïù¥ÌÑ∞ÏÖã Ï§ë ÌïòÎÇòÏùò Ïù¥ÎØ∏ÏßÄÎ•º ÏãúÍ∞ÅÌôîÌï¥Î≥¥Í∏∞
import matplotlib.pyplot as plt
import numpy as np

# Îç∞Ïù¥ÌÑ∞ÏÖãÏóêÏÑú ÌïòÎÇòÏùò ÏÉòÌîå Í∞ÄÏ†∏Ïò§Í∏∞
image, label = train_subset[1]

# ÌÖêÏÑúÎ•º NumPy Î∞∞Ïó¥Î°ú Î≥ÄÌôò
image = image.numpy().squeeze()

# Ïù¥ÎØ∏ÏßÄ Ï∂úÎ†•
plt.imshow(image)  # Ïù¥ÎØ∏ÏßÄ ÌëúÏãú
plt.title(f"Label: {label}")  # Î†àÏù¥Î∏î ÌëúÏãú
plt.axis("off")  # Ï∂ï ÏóÜÏï†Í∏∞
plt.show()
```

## **1. ANN**

**Ïù∏Í≥µ Ïã†Í≤ΩÎßù (ANN)**

>  Î°úÏßÄÏä§Ìã± ÌöåÍ∑ÄÎäî Î∂ÑÎ•ò Î¨∏Ï†úÏóêÏÑú ÏÑ±Îä•Ïù¥ Ï¢ãÏßÄÎßå, Îç∞Ïù¥ÌÑ∞Ïùò ÎπÑÏÑ†ÌòïÏÑ±Ïù¥ Ï¶ùÍ∞ÄÌïòÎ©¥ Î™®Îç∏Ïùò Ï†ïÌôïÎèÑÍ∞Ä Í∞êÏÜåÌï©ÎãàÎã§.

>  Îî∞ÎùºÏÑú, hidden lyaer Ïóê Îçî ÎßéÏùÄ ÎπÑÏÑ†Ìòï Ìï®ÏàòÎ•º Ï∂îÍ∞ÄÏãúÏºú Î™®Îç∏Ïùò Î≥µÏû°ÎèÑÎ•º Ï¶ùÍ∞ÄÏãúÏºú Î¥ÖÏãúÎã§. üòÄ

‚Ü™ **3Í∞úÏùò hidden layerÎ•º ÏåìÍ≥† Îã§ÏñëÌïú ÌôúÏÑ±Ìôî Ìï®Ïàò ReLU, Tanh, ELU ÏÇ¨Ïö©Ìï¥Î¥ÖÏãúÎã§**


```python
# ÌïÑÏöîÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏûÑÌè¨Ìä∏
import torch
import torch.nn as nn
```

#### **ANN Î™®Îç∏ ÏÉùÏÑ±**

```python
# ANN Model
class ANNModel(nn.Module):

    def __init__(self, input_dim, hidden_dim, output_dim):
        super(ANNModel, self).__init__()

        # Linear function 1: input_dim --> hidden_dim  (ÏûÖÎ†• Ï∞®ÏõêÏóêÏÑú Ïà®Í≤®ÏßÑ Í≥ÑÏ∏µÏúºÎ°ú)
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        # Non-linearity 1 (ReLU ÌôúÏÑ±Ìôî Ìï®Ïàò)
        self.relu1 = nn.ReLU()

        # Linear function 2: hidden_dim --> hidden_dim (Ïà®Í≤®ÏßÑ Í≥ÑÏ∏µÏùÑ Ìïú Î≤à Îçî ÌÜµÍ≥º)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        # Non-linearity 2 (Tanh ÌôúÏÑ±Ìôî Ìï®Ïàò ÏÇ¨Ïö©)
        self.tanh2 = nn.Tanh()

        # Linear function 3: hidden_dim --> hidden_dim  (Ï∂îÍ∞Ä Ïà®Í≤®ÏßÑ Í≥ÑÏ∏µ)
        self.fc3 = nn.Linear(hidden_dim, hidden_dim)
        # Non-linearity 3 (ELU ÌôúÏÑ±Ìôî Ìï®Ïàò)
        self.elu3 = nn.ELU()

        # Linear function 4: hidden_dim --> output_dim (ÎßàÏßÄÎßâ Ï∂úÎ†•Ï∏µ)
        self.fc4 = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        # Linear function 1
        out = self.fc1(x)
        # Non-linearity 1
        out = self.relu1 (out)

        # Linear function 2
        out = self.fc2(out)
        # Non-linearity 2
        out = self.tanh2 (out)

        # Linear function 3
        out = self.fc3(out)
        # Non-linearity 3
        out = self.elu3 (out)

        # Linear function 4
        out = self.fc4(out)  # ÎßàÏßÄÎßâ Ï∂úÎ†•Ï∏µ
        return out
```

```python
# ANN Í∞ùÏ≤¥ ÏÉùÏÑ±
input_dim = 28 * 28  # Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ (28x28)
hidden_dim = 150  # ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ (150ÏúºÎ°ú ÏÑ§Ï†ï)
output_dim = 10   # Ï∂úÎ†• ÌÅ¥ÎûòÏä§ Í∞úÏàò (0~4)


model = ANNModel(input_dim,hidden_dim,output_dim)
# ÏÜêÏã§ Ìï®Ïàò ÏÑ§Ï†ï
loss_ftn = nn.CrossEntropyLoss() # Îã§Ï§ëÎ∂ÑÎ•òÏóê Ï†ÅÌï©Ìïú ÏÜêÏã§Ìï®ÏàòÎ•º ÏÑ†ÌÉùÌï¥Ï£ºÏÑ∏Ïöî

# SGD Optimizer ÏÇ¨Ïö© - Îã§Î•∏ ÏòµÌã∞ÎßàÏù¥Ï†ÄÎ•º ÏÇ¨Ïö©Ìï¥Î≥¥ÏÖîÎèÑ Îê©ÎãàÎã§ :)
learning_rate = 0.02
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

#### **ANN Î™®Îç∏ ÌïôÏäµ**

```python
# ANN Î™®Îç∏ ÌïôÏäµ

# ÌïôÏäµÏùÑ Î™®ÎãàÌÑ∞ÎßÅÌïòÍ∏∞ ÏúÑÌïú Î≥ÄÏàòÎì§
count = 0  # Î∞òÎ≥µ ÌöüÏàò
loss_list = []  # Loss Ï†ÄÏû• Î¶¨Ïä§Ìä∏
iteration_list = []  # Iteration Ï†ÄÏû• Î¶¨Ïä§Ìä∏
accuracy_list = []  # Ï†ïÌôïÎèÑ Ï†ÄÏû• Î¶¨Ïä§Ìä∏

# Ï†ÑÏ≤¥ ÌïôÏäµ
for epoch in range(10):

    # Î∞∞Ïπò Îã®ÏúÑ ÌïôÏäµ
    for i, (images, labels) in enumerate(train_loader):

        # Îç∞Ïù¥ÌÑ∞Î•º Î™®Îç∏ ÏûÖÎ†• ÌòïÌÉúÎ°ú Î≥ÄÌôò (2D ÌÖêÏÑúÎ°ú ÌéºÏπòÍ∏∞)
        train = images.view(-1, 28*28)  # 28x28 Ïù¥ÎØ∏ÏßÄÎ•º ÌéºÏ≥êÏÑú 1D Î≤°ÌÑ∞Î°ú Î≥ÄÌôò
        labels = labels  # ÌÖêÏÑú ÌòïÌÉú Ïú†ÏßÄ

        # ÏòµÌã∞ÎßàÏù¥Ï†ÄÏùò Í∏∞Ïö∏Í∏∞ Ï¥àÍ∏∞Ìôî
        optimizer.zero_grad()

        # Forward
        # Î™®Îç∏ÏùÑ Ïù¥Ïö©Ìï¥ ÏòàÏ∏°Í∞í(outputs) Í≥ÑÏÇ∞
        outputs = model(train)

        # ÏÜêÏã§ Í≥ÑÏÇ∞
        # cross entropy lossÎ•º ÏÇ¨Ïö©Ìï¥ ÏòàÏ∏°Í∞íÍ≥º Ïã§Ï†úÍ∞í ÎπÑÍµê
        loss = loss_ftn(outputs, labels)

        # Backward propagation - ÏÜêÏã§Ïóê ÎåÄÌïú Í∏∞Ïö∏Í∏∞ Í≥ÑÏÇ∞
        loss.backward()

        # Í∞ÄÏ§ëÏπò ÏóÖÎç∞Ïù¥Ìä∏
        optimizer.step()

        count += 1  # Î∞òÎ≥µ ÌöüÏàò Ï¶ùÍ∞Ä

        if count % 50 == 0:
            # Ï†ïÌôïÎèÑ Í≥ÑÏÇ∞ÏùÑ ÏúÑÌïú Î≥ÄÏàò Ï¥àÍ∏∞Ìôî
            correct = 0
            total = 0


            # Î™®Îç∏Ïù¥ ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ÏÖãÏóêÏÑú ÏñºÎßàÎÇò Ïûò ÏòàÏ∏°ÌïòÎäîÏßÄ ÌôïÏù∏
            for images, labels in test_loader:

                test = images.view(-1, 28*28)  # 28x28 Ïù¥ÎØ∏ÏßÄÎ•º ÌéºÏ≥êÏÑú 1D Î≤°ÌÑ∞Î°ú Î≥ÄÌôò
                # Forward
                # ÏòàÏ∏°Í∞í Í≥ÑÏÇ∞
                outputs =model(test)

                # ÏòàÏ∏°Í∞í Ï§ë Í∞ÄÏû• ÎÜíÏùÄ Í∞íÏùÑ ÏÑ†ÌÉù
                predicted = torch.max(outputs.data, 1)[1]

                # Ï¥ù ÏÉòÌîå Í∞úÏàò
                total += len(labels)

                # Ïò¨Î∞îÎ•¥Í≤å ÏòàÏ∏°Ìïú Í∞úÏàò
                correct += (predicted == labels).sum()

            # Ï†ïÌôïÎèÑ Í≥ÑÏÇ∞
            accuracy = 100 * correct / float(total)

            # ÏÜêÏã§ Î∞è Ï†ïÌôïÎèÑ Ï†ÄÏû•
            loss_list.append(loss.item())
            iteration_list.append(count)
            accuracy_list.append(accuracy)

        if count % 500 == 0:
            # ÏÜêÏã§ Î∞è Ï†ïÌôïÎèÑ Ï∂úÎ†•
            print('Iteration: {}  Loss: {}  Accuracy: {} %'.format(count, loss.data, accuracy))
```

#### **ANN ÌïôÏäµ Í≤∞Í≥º**

```python
# loss ÏãúÍ∞ÅÌôî
plt.plot(iteration_list,loss_list)
plt.xlabel("Number of iteration")
plt.ylabel("Loss")
plt.title("ANN: Loss vs Number of iteration")
plt.show()

# accuracy ÏãúÍ∞ÅÌôî
plt.plot(iteration_list,accuracy_list,color = "red")
plt.xlabel("Number of iteration")
plt.ylabel("Accuracy")
plt.title("ANN: Accuracy vs Number of iteration")
plt.show()
```

#### **ANN ÏµúÏ†ÅÏùò ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ Ï∞æÍ∏∞**

1Ô∏è‚É£ **Ïã§Ìóò 1: Hidden Dimension ÏµúÏ†ÅÍ∞í Ï∞æÍ∏∞**
- hidden_dim Í∞íÏùÑ 50-200 ÏÇ¨Ïù¥ Í∞íÎì§Î°ú ÏÑ§Ï†ïÌïòÍ≥† Î™®Îç∏ÏùÑ ÌïôÏäµ
- Í∞Å hidden_dim Í∞íÏóê ÎåÄÌï¥ ÌïôÏäµÏùÑ ÏàòÌñâÌïòÎ©¥ÏÑú lossÍ≥º accuracyÎ•º Ï†ÄÏû•
- Î™®Îì† Í≤∞Í≥ºÎ•º Ìïú Í∑∏ÎûòÌîÑÏóê ÌëúÏãúÌïòÏó¨ ÎπÑÍµê




2Ô∏è‚É£ **Ïã§Ìóò 2: Learning Rate ÏµúÏ†ÅÍ∞í Ï∞æÍ∏∞**

- learning_rate Í∞íÏùÑ 0.1-0.005 ÏÇ¨Ïù¥ Í∞íÎì§Î°ú ÏÑ§Ï†ïÌïòÍ≥† Î™®Îç∏ÏùÑ ÌïôÏäµ
- Í∞Å learning_rate Í∞íÏóê ÎåÄÌï¥ ÌïôÏäµÏùÑ ÏàòÌñâÌïòÎ©¥ÏÑú lossÍ≥º accuracyÎ•º Ï†ÄÏû•
- Î™®Îì† Í≤∞Í≥ºÎ•º Ìïú Í∑∏ÎûòÌîÑÏóê ÌëúÏãúÌïòÏó¨ ÎπÑÍµê



**‚ö† Ïù¥Îïå epoch Í∞íÏùÄ ÏûêÏú†Î°≠Í≤å Ìï¥Ï£ºÏÖîÎèÑ Îê©ÎãàÎã§! **

```python
# Ïã§Ìóò 1 hidden_dims=[50-200 ÏÇ¨Ïù¥ Í∞íÎì§] Î°ú ÌïôÏäµ

hidden_dims = [50, 100, 150, 200]
num_epochs = 5
learning_rate = 0.02

results_hidden = {}

for hidden_dim in hidden_dims:

    model = ANNModel(input_dim=28*28, hidden_dim=hidden_dim, output_dim=10)
    loss_ftn = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

    loss_list = []
    accuracy_list = []

    for epoch in range(num_epochs):
        for images, labels in train_loader:
            train = images.view(-1, 28*28)
            labels = labels

            optimizer.zero_grad()
            outputs = model(train)
            loss = loss_ftn(outputs, labels)
            loss.backward()
            optimizer.step()

            loss_list.append(loss.item())


        correct, total = 0, 0
        for images, labels in test_loader:
            test = images.view(-1, 28*28)
            outputs = model(test)
            predicted = torch.max(outputs.data, 1)[1]
            total += len(labels)
            correct += (predicted == labels).sum().item()
        accuracy_list.append(100 * correct / total)

    results_hidden[hidden_dim] = (loss_list, accuracy_list)
```

```python
# Ïã§Ìóò 1 Î™®Îç∏ ÌïôÏäµ Í≤∞Í≥º ÏãúÍ∞ÅÌôî
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))


plt.subplot(1, 2, 1)
for hd in hidden_dims:
    plt.plot(results_hidden[hd][0], label=f'Hidden Dim {hd}')
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Loss vs Iterations (Hidden Dim)")
plt.legend()


plt.subplot(1, 2, 2)
for hd in hidden_dims:
    plt.plot(results_hidden[hd][1], label=f'Hidden Dim {hd}')
plt.xlabel("Iterations")
plt.ylabel("Accuracy (%)")
plt.title("Accuracy vs Iterations (Hidden Dim)")
plt.legend()

plt.tight_layout()
plt.show()
```

ÏûëÏùÄ hidden dimensionÏùÄ ÌïôÏäµÌï† Ïàò ÏûàÎäî Ìå®ÌÑ¥Ïù¥ Ï†úÌïúÎêòÏñ¥ Ï†ïÌôïÎèÑÍ∞Ä ÎÇÆÏùÑ Ïàò ÏûàÎã§. Î∞òÎåÄÎ°ú ÎÑàÎ¨¥ ÌÅ∞ hidden dimensionÏùÄ ÌïôÏäµ Ïö©ÎüâÏùÄ Ï∂©Î∂ÑÌïòÏßÄÎßå Í≥ºÏ†ÅÌï© ÏúÑÌóòÏù¥ Ïª§ÏßÑÎã§. Ï¶â, hidden dimensionÏùÄ Î™®Îç∏ Î≥µÏû°ÎèÑÏôÄ ÏùºÎ∞òÌôî ÏÑ±Îä• ÏÇ¨Ïù¥Ïùò Í∑†ÌòïÏùÑ Ï°∞Ï†àÌïòÎäî Ï§ëÏöîÌïú ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞Ïù¥Îã§.

```python
# Ïã§Ìóò 2 learning_rates=[0.1-0.005 ÏÇ¨Ïù¥ Í∞íÎì§] Î°ú ÌïôÏäµ
learning_rates = [0.1, 0.05, 0.02, 0.005]
hidden_dim = 150
num_epochs = 5

results_lr = {}

for lr in learning_rates:

    model = ANNModel(input_dim=28*28, hidden_dim=hidden_dim, output_dim=10)
    loss_ftn = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=lr)

    loss_list = []
    accuracy_list = []

    for epoch in range(num_epochs):
        for images, labels in train_loader:
            train = images.view(-1, 28*28)
            labels = labels

            optimizer.zero_grad()
            outputs = model(train)
            loss = loss_ftn(outputs, labels)
            loss.backward()
            optimizer.step()

            loss_list.append(loss.item())


        correct, total = 0, 0
        for images, labels in test_loader:
            test = images.view(-1, 28*28)
            outputs = model(test)
            predicted = torch.max(outputs.data, 1)[1]
            total += len(labels)
            correct += (predicted == labels).sum().item()
        accuracy_list.append(100 * correct / total)

    results_lr[lr] = (loss_list, accuracy_list)
```

```python
# Ïã§Ìóò 2 Î™®Îç∏ ÌïôÏäµ Í≤∞Í≥º ÏãúÍ∞ÅÌôî
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
for lr in learning_rates:
    plt.plot(results_lr[lr][0], label=f'LR {lr}')
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Loss vs Iterations (Learning Rate)")
plt.legend()

plt.subplot(1, 2, 2)
for lr in learning_rates:
    plt.plot(results_lr[lr][1], label=f'LR {lr}')
plt.xlabel("Iterations")
plt.ylabel("Accuracy (%)")
plt.title("Accuracy vs Iterations (Learning Rate)")
plt.legend()

plt.tight_layout()
plt.show()
```

## **2. CNN**
> CNNÏùÄ Ïù¥ÎØ∏ÏßÄ Î∂ÑÎ•òÏóê Ï†ÅÌï©Ìïú Î™®Îç∏Î°ú, kernelÏùÑ Ïù¥Ïö©Ìï¥ feature mapÏùÑ ÏÉùÏÑ±Ìï©ÎãàÎã§

**‚Ü™ 2Í∞úÏùò Ìï©ÏÑ±Í≥± Í≥ÑÏ∏µÏùÑ ÏåìÍ≥†, ÌôúÏÑ±Ìôî Ìï®ÏàòÎ°ú ReLUÎ•º ÏÇ¨Ïö©ÌïòÎ©∞,Max Pooling layerÎ•º Ï†ÅÏö©Ìï¥Î¥ÖÏãúÎã§.**

#### **CNN Î™®Îç∏ ÏÉùÏÑ±**

```python
import torch
import torch.nn as nn
import torch.optim as optim

# CNN Model Ï†ïÏùò
class CNNModel(nn.Module):
    def __init__(self):
        super(CNNModel, self).__init__()

        # Ìï©ÏÑ±Í≥± Í≥ÑÏ∏µ 1 (Conv Layer 1)
        self.cnn1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=5, stride=1, padding=0)
        self.relu1 = nn.ReLU()  # ÌôúÏÑ±Ìôî Ìï®Ïàò (ReLU)

        # Max Pooling 1
        self.maxpool1 = nn.MaxPool2d(kernel_size=2)  # kernel sizeÎ•º 2Î°ú ÏÑ§Ï†ï

        # Ìï©ÏÑ±Í≥± Í≥ÑÏ∏µ 2 (Conv Layer 2)
        self.cnn2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=5, stride=1, padding=0)
        self.relu2 = nn.ReLU()  # ÌôúÏÑ±Ìôî Ìï®Ïàò (ReLU)

        # Max Pooling 2
        self.maxpool2 = nn.MaxPool2d(kernel_size=2) # kernel sizeÎ•º 2Î°ú ÏÑ§Ï†ï

        # Fully Connected Layer
        self.fc1 = nn.Linear(32 * 4 * 4, 10)  # ÏûÖÎ†• Ï∞®Ïõê Î∞è Ï∂úÎ†• Ï∞®Ïõê ÏßÄÏ†ï

    def forward(self, x):
        # Ìï©ÏÑ±Í≥± Í≥ÑÏ∏µ 1 + ÌôúÏÑ±Ìôî Ìï®Ïàò
        out = self.cnn1(x)
        out = self.relu1(out)

        # Max Pooling 1
        out = self.maxpool1(out)

        # Ìï©ÏÑ±Í≥± Í≥ÑÏ∏µ 2 + ÌôúÏÑ±Ìôî Ìï®Ïàò
        out = self.cnn2(out)
        out = self.relu2(out)

        # Max Pooling 2
        out = self.maxpool2(out)

        # feature mapÏùÑ flattenÌïòÍ≤å
        out = out.view(out.size(0), -1)

        # ÏôÑÏ†Ñ Ïó∞Í≤∞ Í≥ÑÏ∏µ
        out = self.fc1(out)

        return out
```

```python
# CNN Î™®Îç∏ ÌïôÏäµ

# ÌïôÏäµÏùÑ Î™®ÎãàÌÑ∞ÎßÅÌïòÍ∏∞ ÏúÑÌïú Î≥ÄÏàòÎì§
count = 0  # Î∞òÎ≥µ ÌöüÏàò
loss_list = []  # Loss Ï†ÄÏû• Î¶¨Ïä§Ìä∏
iteration_list = []  # Iteration Ï†ÄÏû• Î¶¨Ïä§Ìä∏
accuracy_list = []  # Ï†ïÌôïÎèÑ Ï†ÄÏû• Î¶¨Ïä§Ìä∏
num_epochs = 20  # ÌïôÏäµÌï† ÏóêÌè¨ÌÅ¨ Ïàò ÏÑ§Ï†ï -> ÏûêÏú†Î°≠Í≤å ÏÑ§Ï†ïÌï¥Ï£ºÏÖîÎèÑ Îê©ÎãàÎã§

for epoch in range(num_epochs):
    running_loss = 0.0  # ÏóêÌè¨ÌÅ¨Î≥Ñ ÏÜêÏã§ Ï¥àÍ∏∞Ìôî

    for images, labels in train_loader:
        # ÏòµÌã∞ÎßàÏù¥Ï†ÄÏùò Í∏∞Ïö∏Í∏∞ Ï¥àÍ∏∞Ìôî
        optimizer.zero_grad()

        # Forward
        outputs = model(images)

        # ÏÜêÏã§ Í≥ÑÏÇ∞
        loss = loss_ftn(outputs, labels)

        # Backward
        loss.backward()


        # Í∞ÄÏ§ëÏπò ÏóÖÎç∞Ïù¥Ìä∏
        optimizer.step()

        # ÏÜêÏã§ ÎàÑÏ†Å
        running_loss += loss.item()
        count += 1  # Î∞òÎ≥µ ÌöüÏàò Ï¶ùÍ∞Ä

        if count % 50 == 0:
            correct = 0
            total = 0

            with torch.no_grad():  # ÌèâÍ∞Ä
                for images, labels in test_loader:
                    # Î™®Îç∏ ÏòàÏ∏° ÏàòÌñâ
                    outputs = model(images)

                    # Í∞ÄÏû• ÌôïÎ•†Ïù¥ ÎÜíÏùÄ ÌÅ¥ÎûòÏä§ ÏÑ†ÌÉù
                    predicted = torch.max(outputs.data, 1)[1]

                    # Ï¥ù ÏÉòÌîå Ïàò Î∞è Ïò¨Î∞îÎ•∏ ÏòàÏ∏° Í∞úÏàò ÎàÑÏ†Å
                    total += labels.size(0)
                    correct += (predicted == labels).sum().item()

            # Ï†ïÌôïÎèÑ Í≥ÑÏÇ∞
            accuracy = 100 * correct / float(total)

            # ÏÜêÏã§ Î∞è Ï†ïÌôïÎèÑ Ï†ÄÏû•
            loss_list.append(loss.item())
            iteration_list.append(count)
            accuracy_list.append(accuracy)

    # Îß§ ÏóêÌè¨ÌÅ¨ÎßàÎã§ ÌèâÍ∑† ÏÜêÏã§ Ï∂úÎ†•
    print(f"Epoch {epoch+1}, Loss: {running_loss / len(train_loader)}")
```
